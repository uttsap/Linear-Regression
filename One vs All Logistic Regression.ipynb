```python
import scipy.misc
import glob
import numpy as np
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import imageio
from numpy import e
from sklearn.metrics import confusion_matrix
```

```python

# load own handwriting
our_own_dataset = []

# load the png image data
for image_file_name in glob.glob('nhcd/numerals/*/*.jpg'):
    # use the filename to set the correct label
    label = int(image_file_name[-12:-11])
    
    # load image data from png files into an array
    print ("loading ... ", image_file_name)
    img_array = scipy.misc.imread(image_file_name, flatten=True)  #Read an image from a file as an array
    
    # reshape from 28x28 to list of 784 values, invert values
    img_data  = (255.0 - img_array.reshape(784))/255.0
    
    # then scale data to range from 0.01 to 1.0
#     img_data = (img_data / 255.0 * 0.99) + 0.01
#     print(np.min(img_data))
#     print(np.max(img_data))
    
    # append label and image data  to test data set
    record = np.append(label,img_data)
    our_own_dataset.append(record)
```

```python
data = np.array(our_own_dataset)
np.random.shuffle(data)
data_pd = pd.DataFrame(data)
data_pd.head()
```

```python
X = data_pd.iloc[:,1:]
ones = np.ones([len(X),1])
X = np.concatenate((ones,X), axis = 1)
y = data_pd.iloc[:,0:1].values
X_train,X_test,y_train,y_test =  train_test_split(X,y,test_size=0.2)
X_train,X_validate,y_train,y_validate = train_test_split(X_train,y_train,test_size=0.2)
```

```python
def sigmoid(z):
# Computes the Sigmoid function of z
    g = 1.0 / (1.0 + e ** (-1.0 * z))
    return g 
```

```python
def hypothesis(w, X):
    h = sigmoid(X @ w.T)
    return h
```

```python
def calCost(w,X,y):
    h = sigmoid(X @ w.T)
    N = len(X)
    cost = -((1/N) * np.sum((y * np.log(h)) + (1-y)*np.log(1-h)))
    return cost
```

```python
def computeGradient(w,X,y):
    h = sigmoid(X @ w.T)
    N = len(X)
    gradient = ((1/N) * (np.sum(X*(h-y),axis = 0)))
    return gradient
```

```python
def updateWeight(w,X,y,a=0.03,epochs=100):
    for i in range(epochs):
            w = w - (a * computeGradient(w,X,y))
            cost = calCost(w,X,y)
            print('Cost',cost)
    return w, cost
```

```python
def precision(label, confusion_matrix):
    col = confusion_matrix[:, label]
    return confusion_matrix[label, label] / col.sum()
```

```python
def recall(label, confusion_matrix):
    row = confusion_matrix[label, :]
    return confusion_matrix[label, label] / row.sum()
```

```python
print("label precision recall")
for label in range(10):
    print(f"{label:5d} {precision(label, cm):9.3f} {recall(label, cm):6.3f}")
```

```python
def precision_macro_average(confusion_matrix):
    rows, columns = confusion_matrix.shape
    sum_of_precisions = 0
    for label in range(rows):
        sum_of_precisions += precision(label, confusion_matrix)
    return sum_of_precisions / rows
def recall_macro_average(confusion_matrix):
    rows, columns = confusion_matrix.shape
    sum_of_recalls = 0
    for label in range(columns):
        sum_of_recalls += recall(label, confusion_matrix)
    return sum_of_recalls / columns
```

```python
print("precision total:", precision_macro_average(cm))
print("recall total:", recall_macro_average(cm))
```

```python
def accuracy(confusion_matrix):
    diagonal_sum = confusion_matrix.trace()
    sum_of_all_elements = confusion_matrix.sum()
    return diagonal_sum / sum_of_all_elements 
```
